```python
import pandas as pd
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
```

```python
mm18 = glob.glob(os.path.join('.', '18mm*.txt'))
mm25 = glob.glob(os.path.join('.', '25mm*.txt'))
mm30 = glob.glob(os.path.join('.', '30mm*.txt'))
display(mm18)
display(mm25)
display(mm30)
```

```python
after = []
before = []
i = ["No Block", "Light Block", "Heavy Block", "Complete Block"]
for file in mm18:
    df = pd.read_csv(file, header=None, sep='     ')
    after.append(df.iloc[:, 1][0])
    before.append(df.iloc[:, 1][1])

plt.plot(i, before, label="Before Orrifice Pressure")
plt.plot(i, after, label="After Orrifice Pressure")
plt.legend()
plt.title("18mm Orrifice Measurement")
plt.ylabel("Pressure (Pa)")
plt.savefig("Preliminary 18mm.png")
plt.show()
```
